```python
import pandas as pd
import numpy as np
from fuzzywuzzy import process

# Example column
education_col = train['Education Speaciality']

# Basic cleaning
def normalize_text(text):
    if pd.isna(text) or not isinstance(text, str):
        return np.nan
    text = text.strip().lower()
    text = text.replace("–", "-")  # Normalize dashes
    return text

# Standardized categories
standard_categories = {
    # Computer Science and IT
    "Computer Science and IT":["computer science", "cs", "علوم الحاسب", "علوم حاسب", "علوم الحاسب الآلي", "computer since", "web devolper","database and programming","البرمجه","visual computing","البرمجة","برمجه","التقنيه", "الحوسبة الكمية","دعم فني","inter-iot","computer scince", "علوم حاسبات", "علوم الحاشب", "quantum computing" ,"علوم حاسبب", "علوم الحاسب والمعلومات", "حاسبات", "حاسب", "حاسوب", "علوم الحاسبات", "حاسب الي", "حاسب آلي", "حاسب الالي", "حاسب ألي", "حاسبات آلية", "computer and information technology", "الحاسب وتقنية المعلومات","it", "information technology", "تقنية المعلومات", "تقنيه معلومات", "information technology it", "information technologies", "تقنية معلومات وحاسوب", "تقنية معلومات و حوسبة", "تقنية معلومات it", "تقنية معلومات-الأمن السيبراني", "تقنية المعلومات والحوسبة","software engineering", "software engineer", "software development", "هندسة برمجيات", "هندسه برمجيات", "software system engineering", "هندسة البرمجيات", "برمجيات", "برمجة حاسب", "برمجة حاسب الي", "برمجة الحاسب الألي", "برمجة الحاسب الالي", "برمجة الحاسب ألي", "برمجة وقواعد بيانات", "برمجة وتطوير الويب", "تطوير التطبيقات","cybersecurity", "امن سيبراني", "أمن المعلومات", "cyber security", "security", "الامن السيبراني", "information security", "أمن سيبراني", "الامن السبراني", "cybersecurity engineering", "هندسة امن المعلومات", "cybersecurity and info assurance", "امن المعلومات والشبكات", "الشبكات وامن البيانات", "الأمن السيبراني والتحري الرقمي","mis", "management information systems", "information systems", "نظم المعلومات", "نظم معلومات", "نظم معلومات ادارية", "نظم المعلومات الإدارية", "نظم معلومات حاسوبية", "نظم المعلومات الحاسوبية", "نظم معلومات حاسب", "نظم المعلومات الادارية", "نظم معلومات إدارية", "نظم المعلومات الإدارية mis", "نظم معلومات حاسوبيه", "نظم المعلومات الجغرافية", "نظم المعلومات التطبيقية","شبكات", "network", "networking", "network engineer", "network security engineering", "شبكات الحاسب", "شبكات الحاسب الآلي", "شبكات الحاسبات", "شبكات الحاسوب", "computer networking", "computer networks", "هندسة شبكات", "هندسة الشبكات", "هندسة الشبكات والاتصالات", "شبكات وامن معلومات", "شبكات وانظمة اتصالات", "شبكات الحاسب و الاتصالات","computer engineering", "هندسة حاسب", "هندسة الحاسب", "هندسة حاسوب", "هندسة الحاسوب", "هندسة كمبيوتر", "هندسة حاسب الي", "هندسة حاسب آلي", "هندسة حاسب الالي", "هندسة الحاسب الآلي", "هندسة الحاسب الالي", "هندسة حاسبات", "هندسة حاسبات آلية", "computer systems engineering", "هندسة النظم والحاسبات","data science", "علم البيانات", "علوم البيانات", "تحليل البيانات", "data analytics", "data management", "علم البيانات وتحليلها", "علوم البيانات وتحليلها", "نظم المعلومات - علم بيانات", "نظم معلومات مسار علم بيانات", "big data","artificial intelligence", "الذكاء الاصطناعي", "ai", "ذكاء اصطناعي"],

    # Business and Management
    "Business and Management":["ادارة اعمال", "اداره عامه","سكرتير تنفيذي","تأمين واداة مخاطر","إقتصاد","الأنظمة","إدارة أعمال","تجارة الكترونية","e-businesses","تجارة إلكترونية" ,"قيادة وتخطيط", "اقتصاد كلية الاداره والاعمال","pmp","business administration", "mba", "اداره اعمال", "إدارة إعمال", "إدارة الاعمال", "ادارة الأعمال", "اقتصاد","business management", "ادارة الاعمال الالكترونية", "business and adminstration","management information system", "mis", "management information systems", "نظم المعلومات الإدارية", "نظم معلومات ادارية", "نظم المعلومات الادارية", "نظم معلومات إدارية", "نظم المعلومات الاداريه", "نظم معلومات اداريه","human resources", "الموارد البشرية", "موارد بشرية", "ادارة الموارد البشرية", "إدارة الموارد البشرية", "human resources management", "ادارة موارد بشرية", "إدارة موارد بشرية", "الموارد البشرية والتطوير","finance", "التمويل", "المالية", "التمويل والاستثمار", "تمويل واستثمار", "العلوم المالية", "المالية والبنوك", "الاعمال المصرفية","accounting", "المحاسبة", "محاسبة", "controlling and accounting","marketing", "التسويق", "تسويق", "الاعلان والاتصال التسويقي", "إعلان واتصال تسويقي", "اتصال استراتيجي", "public relations", "العلاقات العامة"],

    # Media and Communication
    "Media and Communication":["media", "الإعلام", "إعلام", "الاعلام", "إعلام رقمي", "media studies", "دراسات إعلامية", "communication", "الاتصال","journalism", "صحافة", "الصحافة", "صحافة والنشر الإلكتروني","الاعلام (صحافة)", "إعلام (صحافة)", "journalism and media","radio and television", "الإذاعة والتلفزيون", "إذاعة وتلفزيون","الإذاعة والتلفزيون والفيلم", "radio tv and film", "تلفزيون وإذاعة", "الاعلام المرئي", "media production","public relations", "العلاقات العامة", "علاقات عامة", "relations publiques", "pr", "الاعلام والعلاقات العامة", "الاتصال والعلاقات العامة", "strategic communication","digital media", "الإعلام الرقمي", "إعلام رقمي","الوسائط الرقمية", "وسائط متعددة تفاعلية", "digital content", "المحتوى الرقمي","advertising", "إعلان", "الإعلان", "التسويق الإعلاني","الاعلان والاتصال التسويقي", "إعلان واتصال تسويقي","creative advertising", "الإعلان الإبداعي","film production", "إنتاج أفلام", "صناعة الأفلام","الفيلم والسينما", "cinema production", "إخراج أفلام","media management", "إدارة الإعلام", "إدارة مؤسسات إعلامية","إدارة وسائل الإعلام", "media administration"],

    # Engineering
    "Engineering":["electrical engineering", "هندسة كهربائية", "هياكل ومحركات","هندسة كهرباء", "صيانة الالات الصناعية","هندسة الكهرباء", "الهندسة الكهربائية", "آلات دقيقة وتحكم","هندسة الكترونيات", "هندسة إلكترونيات", "هندسة الكترونيات واتصالات", "electronics engineering", "هندسة اتصالات", "هندسة الاتصالات","mechanical engineering", "هندسة ميكانيكية", "كلية المدربين التقنين","هندسه ميكانيكية", "الهندسة الميكانيكية", "هندسة الميكانيكا", "الكلية التقنية","mechanical engineer technology","civil engineering", "هندسة مدنية", "هندسة الإنشاءات", "civil and environmental engineering","industrial engineering", "هندسة صناعية", "هندسه صناعيه", "الهندسة الصناعية", "industrial and system engineering","chemical engineering", "هندسة كيميائية", "الهندسة الكيميائية","biomedical engineering", "هندسة طبية حيوية", "الهندسة الطبية الحيوية", "تكنولوجيا طبية حيوية"],

    # Health Sciences
    "Health Sciences":["medicine", "الطب", "الطب و الجراحة العامة", "دكتور صيدلي","respiratory care","الإعاقة العقلية","امراض معدية","علاج وظيفي" ,"radiology technologist" ,"اشعه تشخيصيه","الاستشارات الوراثية","medical", "طب", "dental surgery", "جراحة الأسنان","pharmacy", "صيدلة", "علم الصيدلة","nursing", "تمريض", "التمريض","public health", "الصحة العامة", "الصحة المجتمعية","health informatics", "المعلوماتية الصحية", "معلوماتيه صحية", "تقنية وادارة المعلومات الصحية", "health information systems"],

    # Sciences
    "Sciences": ["physics", "الفيزياء", "فيزياء", "العلوم الفيزيائية", "جيولوجيا","فيزياء عامة","بحوث العمليات","مختبرات كيميائية","تحضيري علمي","bsc geology","actuarial science","chemistry", "الكيمياء", "كيمياء", "الكيمياء الحيوية","كيمياء حيوية", "biology", "الاحياء", "أحياء", "علم الأحياء", "علوم الحياة", "mathematics", "الرياضيات", "رياضيات", "الرياضيات التطبيقية","رياضيات تطبيقية", "الرياضيات والاحصاء", "رياضيات واحصاء","statistics", "الإحصاء", "احصاء", "علم الإحصاء", "الاحصاء التطبيقي", "احصاء تطبيقي"],

    # Humanities and Social Sciences
    "Humanities and Social Sciences" : ["psychology","psychology", "علم النفس", "العلوم النفسية","occupational therapy", "gis","حقوق","المحاماة","السياحة","applied linguistics" ,"التوجية والاصلاح الاسري","الحسبة والرقابة","الادلة الجنائية","اقتصاد منزلي","libraries and information","sociology", "علم الاجتماع", "قانون","العلوم الاجتماعية", "اجتماعيات","خدمة اجتماعية", "خدمة اجتماعيه", "history", "التاريخ", "تاريخ", "تاريخ اسلامي", "الدراسات التاريخية","geography", "الجغرافيا", "جغرافيا", "الجغرافيا البشرية","islamic studies", "الدراسات الإسلامية", "دراسات اسلامية","الدراسات الاسلامية", "القرأن الكريم والدراسات الاسلاميه","أصول الدين","الشريعة","العقيدة والمذاهب المعاصرة","اصول دين","دراسات وقضايا معاصرة"],

    # Arts and Design
    "Art and Design" : ["graphic design", "التصميم الجرافيكي","تصوير ضوئي", "فنون بصرية","الجرافيكس والوسائط المتعددة", "جرافيك ووسائط رقمية", "تصميم جرافيكي ووسائط متعددة","تجميل","beauty","نحت، فنون بصرية","فنون بصرية ( نحت )" , "تربية فنية","interior design", "التصميم الداخلي", "تصميم داخلي", "fashion design", "تصميم الازياء", "تصميم ازياء", "تصميم الأزياء","multimedia", "الوسائط المتعددة", "وسائط متعددة", "الجرافيك والوسائط المتعددة"],

    # Education :
    "Education": ["education","رياض اطفال", "المناهج وطرق التدريس","مناهج وطرق تدريس","التربية", "تعليم", "العلوم التربوية", "تربية خاصة", "تربية فكريه","التدخل المبكر","english literature","التربية الخاصة","english language", "اللغة الانجليزية", "اللغة الإنجليزية", "لغة انجليزية", "آداب إنجليزي","اللغة الانجليزية وادابها","الأدب الإنجليزي","لغات وترجمه اللغه الانجليزيه","arabic language", "اللغة العربية", "لغة عربية", "الآداب العربية","translation", "الترجمة", "ترجمة", "اللغات والترجمة", "الترجمة الفورية"],


    # Special cases
    "other": ["architecture", "العمارة", "هندسة معمارية", "عمارة", "التصميم المعماري","انشاءات معمارية","aviation", "الطيران", "هندسة طيران", "هندسة الفضاء","marine science", "العلوم البحرية", "الهندسة البحرية", "مساحة بحرية","environmental science", "العلوم البيئية", "هندسة البيئة","غير متخصص", "غير محدد", "none",np.nan,"not specified", "nan","أخرى", "other", "مختلف", "various","جامعة الملك سعود","long island university-new york","university of dubuque"]
}


# Invert the mapping for reverse lookup
flat_mapping = {}
for category, aliases in standard_categories.items():
    for alias in aliases:
        flat_mapping[alias] = category

def map_to_standard(text):
    if pd.isna(text):
        return np.nan
    cleaned = normalize_text(text)
    if cleaned in flat_mapping:
        return flat_mapping[cleaned]

    # Fuzzy match as fallback
    best_match, score = process.extractOne(cleaned, list(flat_mapping.keys()))
    if score >= 80:
        return flat_mapping[best_match]

    return cleaned  # Return normalized but unmapped for review
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler


categorical_cols = X.select_dtypes(include=['object', 'bool']).columns
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns


preprocessor = ColumnTransformer(transformers=[
    ('num', StandardScaler(), numerical_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression,SGDClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import StackingClassifier
```

```python
model1 = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('model', RandomForestClassifier(random_state=42,max_depth=11,class_weight='balanced'))
])
model1.fit(X_train, y_train)
p_train1 = model1.predict(X_train)
print("train Accuracy:", accuracy_score(y_train,p_train1))
y_pred1 = model1.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred1))
print("F1 Score:", f1_score(y_test, y_pred1))
print("Precision:", precision_score(y_test, y_pred1))
print("Recall:", recall_score(y_test, y_pred1))
```
